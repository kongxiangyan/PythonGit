## 第二步 爬取贴吧一个页面中所有帖子的数据

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('http://tieba.baidu.com/f?kw=%E4%B8%AD%E5%9B%BD%E4%BC%A0%E5%AA%92%E5%A4%A7%E5%AD%A6&ie=utf-8&pn=0') 
html = response.text

bs = BeautifulSoup(html)
posts = bs.select('div.t_con')

count = 1
post_url_prefix = 'http://tieba.baidu.com'

for post in posts:
    print(count)
    count += 1

    threadlist_title = post.select('.threadlist_title')
    threadlist_title = threadlist_title[0].text.strip()
    print('threadlist_title:'+ threadlist_title)

    url = post.select('a.j_th_tit')
    url = post_url_prefix + url[0].get('href').strip()
    print('url:'+ url)

    author_name = post.select('.frs-author-name')
    author_name = author_name[0].text.strip()
    print('author_name:'+ author_name)

    with open('3.2_one_page.txt', 'a', encoding='utf-8') as fin:
        fin.write(threadlist_title+','+url+','+author_name+'\n')
    print('=======================================================================================================================================')
```

## 第三步 爬取贴吧中所有帖子数据

```python
import requests
from bs4 import BeautifulSoup

# 这一段代码是用来获取本贴吧最后一页的页码。
response = requests.get('http://tieba.baidu.com/f?kw=%E4%B8%AD%E5%9B%BD%E4%BC%A0%E5%AA%92%E5%A4%A7%E5%AD%A6&ie=utf-8&pn=0') 
html = response.text
bs = BeautifulSoup(html)
href = bs.select('.last')[0]['href']
last_page_num = href[href.rfind('=')+1:]


count = 1

# 开始循环所有页
for page in range(0,int(last_page_num)+1,50):
    response = requests.get('http://tieba.baidu.com/f?kw=%E4%B8%AD%E5%9B%BD%E4%BC%A0%E5%AA%92%E5%A4%A7%E5%AD%A6&ie=utf-8&pn='+str(page)) 
    
    html = response.text

    bs = BeautifulSoup(html)
    posts = bs.select('div.t_con')
    
    post_url_prefix = 'http://tieba.baidu.com'

    # 开始循环每一个帖子
    for post in posts:
        print(count)
        count += 1

        reply_num = post.select('.threadlist_rep_num')
        reply_num = reply_num[0].text.strip()
        print('reply_num:'+ reply_num)

        threadlist_title = post.select('.threadlist_title')
        threadlist_title = threadlist_title[0].text.strip()
        print('threadlist_title:'+ threadlist_title)

        url = post.select('a.j_th_tit')
        url = post_url_prefix + url[0].get('href').strip()
        print('url:'+ url)

        author_name = post.select('.frs-author-name')
        author_name = author_name[0].text.strip()
        print('author_name:'+ author_name)

        with open('3.2_all_page.txt', 'a', encoding='utf-8') as fin:
            fin.write(threadlist_title+','+url+','+reply_num+','+author_name+'\n')
        print('=======================================================================================================================================')
```
